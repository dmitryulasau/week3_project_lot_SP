# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
- leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
from IPython.display import clear_output

class Garage():
    def __init__(self):
        self.parking_spaces = 5
#         self.tickets = list(range(1, self.parking_spaces + 1))
        self.current_ticket = {}
        self.ticket = 0
        self.tickets_printed = []
        self.paid_tickets = []
    
    def park(self):
        """Make a parking spot occupied"""
        self.ticket += 1
        self.parking_spaces -=1
        self.make = input("What is your car MAKE? ")
#         self.model = input("What is your car MODEL? ")
#         self.color = input("What is your car COLOR? ")
#         self.plate = input("What is your car NUMBER PLATE? ")
        self.payment = "NO"
        
        self.current_ticket[self.ticket] = {
                'make' : self.make, 
#                 'model': self.model, 
#                 'color': self.color, 
#                 'number_plate' : self.plate,
                'payment': self.payment
            }
        
        self.tickets_printed.append(self.ticket)
        
    def paid(self):
        """Change status of the PAYMENT"""
        for ticket, value in self.current_ticket.items():
            if ticket == self.ticket:
                if self.current_ticket[int(ticket)]["payment"] == "NO":
                    self.current_ticket[int(ticket)]["payment"] = "PAID"
        if self.ticket not in self.paid_tickets:
            self.paid_tickets.append(self.ticket)
  
    def ticket_information(self, number = 0):
        """Print information about ticket"""
        clear_output()
        for ticket, value in self.current_ticket.items():
            if ticket == self.ticket:
                print(f"******************************"
                      f"\nYour ticket is # {self.ticket}:"
                      f"\nMake: {self.make.upper()}"
#                       f"\nModel: {self.model.upper()}"
#                       f"\nColor: {self.color.upper()}"
#                       f"\nNumber plate: {self.plate.upper()}"
                      f"\nPayment: {self.payment}\n"
                      f"******************************\n"
                     )

    def sunset(self):
        """Print Parking logo"""
        print("(*=====[S][U][N][S][E][T]==^==[P][A][R][K][I][N][G]=====*)"
              "\n   ~Please take a ticket before entering a parking lot~   \n"
              f"\nPARKING SLOTS AVAILABLE: -=( {lot.parking_spaces} )=-")
            
###############################################################################################################################       
lot = Garage()

def main():

    while True:

        lot.sunset()
        option = input(                   
                       "\nOPTIONS: "
                       "\n\t- (1) PARK (TAKE A TICKET)"
                       "\n\t- (2) PAY FOR PARKING"
                       "\n\t- (3) LEAVE"
                       "\n\t- (0) QUIT - enter '0' or 'q'\n"

                       "\nWHAT WOULD YOU LIKE? "
                        )
       
        
        # [ WRONG INPUT ] //////////////////////////////////////////////////////////////////////////////////////////+
        if option[0].lower() not in {'1', '2', '3', '0', 'q'} or len(option) > 1:
            clear_output()
            print("\n-* !WRONG INPUT! PLEASE FOLLOW THE INSTRUCTIONS *-\n")
            continue
        
        # [ QUIT ] ////////////////////////////////////////////////////////////////////////////////////////////////+
        if option[0].lower() == 'q' or option == "0":
            clear_output()
            print("\nThank you for visiting us! Have a great day! :) \n"
                  "\n* SUNSET PARKING *"
                    "\n__/‾‾|‾‾\___\n"
                    "'–◯–-––◯–'""")
            break
            
        # [ PARK ] //////////////////////////////////////////////////////////////////////////////////////////////+
        if option == "1":
            clear_output()
            lot.sunset() # Sunset parking LOGO
          
            if lot.parking_spaces != 0:
                
                clear_output()
                print(f"SPACES LEFT - {lot.parking_spaces}")
                print("Please provide some information about your car: ")
                lot.park() # ADDS INFORMATION ABOUT THE CAR TO THE current_ticket {}
                clear_output()
                
                lot.ticket_information() # PRINTS TICKET INFORMATION
                
            # PARKING LOT HAS NO SPACES
            else:
                clear_output()
                print(f"SPACES {lot.parking_spaces}")
                print("\n!*** SORRY. Our parking lot is FULL! ***!\n")
                  
        # [ PAY FOR PARKING ] /////////////////////////////////////////////////////////////////////////////////////
        if option == "2":
            
            # NO TICKETS WERE DISTRIBUTED
            if lot.tickets_printed == []:
                clear_output()
                print("*** NO TICKETS YET ***\n")
                continue
             
            
            clear_output()
            
            print("*** PAYMENT ***")
            print("* 'q' - QUIT")
            print("* 'back' - MAIN MENU\n")
            print(f"Unpaid tickets: {sorted(lot.tickets_printed)} :(")
            print(f"Paid tickets: {sorted(lot.paid_tickets)} :D")

            payment = input("Please enter a ticket number: ")
            
            # QUIT
            if payment[0].lower() == 'q':
                clear_output()
                print("\nThank you for visiting us! Have a great day! :) \n"
                      "\n* SUNSET PARKING *"
                      "\n__/‾‾|‾‾\___\n"
                    "'–◯–-––◯–'""")
                break
            
            # BACK TO THE MAIN MENU
            if payment[0].lower() == 'b':
                clear_output()
                continue
          
            # ENTER A TICKET NUMBER FOR PAYMENT
            if int(payment) in lot.tickets_printed and len(option) == 1:
                
                clear_output() 
                print(f"**You are paying for ticket # {payment}**")
                
                payment_confirmation = input("Would you like to pay? (yes/no) ")
                if payment_confirmation[0].lower() == 'y':
                    clear_output()
                    lot.current_ticket[int(payment)]['payment'] = "PAID"
                    lot.tickets_printed.remove(int(payment))
                    lot.paid_tickets.append(int(payment))

                    print(f"Unpaid tickets: {sorted(lot.tickets_printed)} :(")
                    print(f"Paid tickets: {sorted(lot.paid_tickets)} :D\n")
                    print(f"******************************"
                          f"\nTicket # {payment}: PAID\n"
                          f"THANK YOU FOR YOUR PAYMENT!"
                          "\n*You have 15 minutes to leave.*\n")
                    
                
                else:
                    clear_output()
                    print("\n-* !WRONG INPUT! PLEASE FOLLOW THE INSTRUCTIONS *-\n")
                    continue
            
            else:
                clear_output()
                print("\n-* !WRONG INPUT! PLEASE FOLLOW THE INSTRUCTIONS *-\n")
                continue
                
        # LEAVE
        if option == "3":
            clear_output()
            
            print("*** LEAVING THE GARAGE ***")
            print("For QUIT - enter '0' or 'q'\n")
            print(f"! - Unpaid tickets {sorted(lot.tickets_printed)} - !")
            print(f"PAID tickets {sorted(lot.paid_tickets)} :D - YOU CAN LEAVE NOW! ")
            leave = input("Please insert you ticket: ")
            if leave[0].lower() in {'1', '2', '3', '4', '5', '0', 'q'} or len(option) == 1:
                
                if lot.current_ticket[int(leave)]["payment"] == "PAID":
                    lot.parking_spaces +=1
                    clear_output()
                    lot.paid_tickets.remove(int(leave))
                    
                    lot.ticket = int(leave) -1
                    print("***THANK YOU FOR VISITNIG US!***\n")
                   
                    
                else:
                    clear_output()
                    print(f"!*** PLEASE PAY YOU PARKING ***!\n "
                          f"    !** YOUR TICKET # {leave} **!\n ")
             
if __name__ == "__main__":
    # Driver code
    main()

```
